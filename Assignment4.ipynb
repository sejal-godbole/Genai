```python
from transformers import pipeline
```

```python
def main():
  print("Loading model... Please wait.")
  # Load pre-trained QA pipeline
  qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")


  print("\n=== Simple QA Chatbot ===")
  print("Type 'exit' to quit.\n")


  # Sample context (you can replace this with your own text)
  context = """
  Artificial Intelligence (AI) is the simulation of human intelligence in machines.
  These machines are programmed to think like humans and mimic their actions.
  AI is used in various fields such as healthcare, finance, education, and transportation.
  Machine Learning is a subset of AI that allows systems to learn from data.
  """


  print("Context:")
  print(context)


  while True:
    question = input("\nAsk a question: ")


    if question.lower() == "exit":
      print("Goodbye!")
      break


    # Get answer
    result = qa_pipeline(question=question, context=context)


    # Display result
    print("Answer:", result['answer'])
    print("Confidence:", round(result['score'], 3))
```

```python
if __name__ == "__main__":
  main()
```
